```python
import pandas as pd

# Load cleaned data
df = pd.read_csv("notebooks/preprocess/imdb_clean.csv")
x = df['preprocessed_text'].values
y = df['sentiment'].values  # or the target column name

print(x)
print(y)
```

```python
x_train , x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
import numpy as np

# Replace NaN with empty string
x_train = np.where(pd.isna(x_train), "", x_train)
x_test  = np.where(pd.isna(x_test), "", x_test)

# Convert everything to string
x_train = x_train.astype(str)
x_test  = x_test.astype(str)

# Apply TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
x_train = tfidf.fit_transform(x_train)
x_test  = tfidf.transform(x_test)

```

```python
model =LogisticRegression()
model.fit(x_train,y_train)
```

```python
train_pr=model.predict(x_train)
```

```python
train_acc=accuracy_score(y_train,train_pr)
```

```python
train_acc
```

```python
test_pr=model.predict(x_test)
```

```python
test_acc=accuracy_score(y_test,test_pr)
```
